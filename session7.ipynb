```python
a=5
b=10
try:
    result=a/b
except:
    print("Division by Zero")
else:
    print(result)
finally:
    print("Division Operation")
```

```python
try:
    with open("abc.text","r") as f:
        content=f.read()
except:
    print("File Not Found")
else:
    print(content)
finally:
    print(".......File handling Done")
```
