### Does release year + genre + explicit flag influence popularity?

```python
Year_genre_explicit = df.groupby(['year', 'genre', 'explicit'])['popularity'].mean().reset_index().sort_values(by='year', ascending=False)
Year_genre_explicit.tail()
```

```python
pivot = Year_genre_explicit.pivot_table(index='genre',columns=['year','explicit'],values='popularity')
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=False)
plt.title('Popularity by Genre, Year, and Explicit Flag')
plt.xlabel('Year-Explicit')
plt.ylabel('Genre')
plt.show()
```

***The heatmap shows that genre popularity remains relatively stable across years and explicit content categories, with Pop, Hip-Hop, and R&B consistently maintaining higher popularity levels.***
