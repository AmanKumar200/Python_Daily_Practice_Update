You have marks of students:

marks = [45, 78, 90, 66, 88, 72, 59]
Tasks:  
Convert to NumPy array  
Find average marks  
Find highest & lowest marks  
Find students scoring above average  
Add 5 grace marks to all students

```python
marks = [45, 78, 90, 66, 88, 72, 59]
marks = np.array([45, 78, 90, 66, 88, 72, 59])
print(np.mean(marks))
print("Highest marks is ", np.max(marks),"Lowest marks is ",  np.min(marks))
print(marks+5)
print(marks[marks>np.mean(marks)])
```

Employee salaries:  

salary = [25000, 40000, 35000, 50000, 45000]
Tasks:  
Increase all salaries by 10%  
Find employees earning > 40000   

```python
salary = np.array([25000, 40000, 35000, 50000, 45000])
inc_by_10= salary+0.1*salary
emp = inc_by_10[inc_by_10>40000]
print(inc_by_10)
print(emp)
```

A grayscale image (3x3 pixels):

img = [[100, 150, 200],
       [50, 75, 125],
       [0, 25, 255]]
Tasks:  
Convert to NumPy array  
Increase brightness by 20  
Clip values to max 255  

```python
img = np.array([[100, 150, 200], [50, 75, 125], [0, 25, 255]])
brightness = img+10
clipped = np.clip(brightness, 0, 255)
print(brightness)
print(clipped)
```

Scores:

scores = [55, 65, 75, 85, 95]
Tasks:  
Get students with grades ≥ 75  
Replace scores < 60 with 60  

```python
scores = np.array([55, 65, 75, 85, 95])
grades = scores[scores>=75]
print(grades)
adjusted = np.where(scores < 60, 60, scores)
print(adjusted)
```
