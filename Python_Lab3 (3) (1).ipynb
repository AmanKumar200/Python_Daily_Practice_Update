**Exercise 3:** Open a missing file path and catch `FileNotFoundError` gracefully; print a friendly message.

```python
try:
    with open(r"abc.txt","r") as f:
        print(f.read())
except:
    print("File not found")
else:
    print(f)
finally:
    print("File is not present")
```

**Exercise 5:** From a list `['red','green','blue','yellow']`, randomly shuffle the list and print it. Then sample 2 unique colors.

```python
colors=['red','green','blue','yellow']
random.shuffle(colors)
print(colors)
print(random.choices(colors,k=2))
print(random.sample(colors,2))
```

**Exercise 6:** Simulate 1000 coin flips (`H`/`T`) and print the counts of each.

```python
results = [random.choice(["H","T"]) for i in range(1000)]
Heads = results.count("H")
Tails = results.count("T")
print("Heads",Heads)
print("Tail",Tails)
```

**Exercise 9:** Parse `'2025-09-11 08:30:00'` to a `datetime`, add 90 minutes, and print the result in `YYYY-MM-DD HH:MM` format.

```python
from datetime import datetime, timedelta
s = '2025-09-11 08:30:00'
try:
    dt = datetime.strptime(s, "%Y-%m-%d %H:%M:%S")
except ValueError:
    print("bad input")
else:
    new_dt = dt + timedelta(minutes=90)
    print(new_dt)
```

```python
dte= datetime.strptime('2025-09-11 08:30:00' ,"%Y-%m-%d %H:%M:%S")
print(dte)
future = dte+timedelta(minutes=90)
print(future)
```

**Exercise 10:** Create a small schedule: start at today's date and print the next 5 day names (Mon..Sun) using `datetime`.

```python
from datetime import datetime, timedelta
today = datetime.today()
for i in range(6):
    day = today + timedelta(days=i)
    print(day.strftime("%A"))
```
