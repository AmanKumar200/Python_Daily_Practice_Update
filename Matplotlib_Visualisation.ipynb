```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.plot(x, y, color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("First Graph")

plt.subplot(2,2,2)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("Second Graph")

plt.subplot(2,2,3)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("Third Graph")

plt.subplot(2,2,4)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("Fourth Graph")
```

```python
plt.figure(figsize=(8,8))
plt.subplot(3,1,1)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("First Graph")
plt.grid(True)
plt.subplot(3,1,2)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("Second Graph")
plt.grid(True)
plt.subplot(3,1,3)
plt.plot(x,y,color="blue", linestyle="--", marker="*", linewidth=3, markersize=9)
plt.title("Third Graph")
plt.grid(True)
```

## BAR PLOT

```python
categories = ["A","B","C","D","E"]
values= [5,7,3,8,6]
plt.bar(categories,values,color = ["red","blue","green","yellow","silver"])
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Bar Graph")
```

```python
categories = ["A","B","C","D","E"]
values= [5,7,3,8,6]
plt.barh(categories,values,color = ["red","blue","green","yellow","silver"])
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Bar Graph")
```
